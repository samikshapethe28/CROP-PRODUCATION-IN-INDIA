```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from joblib import dump
```

```python
# Loading the dataset
file_path = r'C:\Users\samik\Downloads\Crop Production data.csv'
df = pd.read_csv(file_path)
```

```python
df.dropna(subset=['Production'], inplace=True)
```

```python
encoder = OrdinalEncoder()
df[['State_Name', 'District_Name', 'Season', 'Crop']] = encoder.fit_transform(df[['State_Name', 'District_Name', 'Season', 'Crop']])
```

```python
# Spliting data into feature variables
X = df.drop(columns=['Production'])
y = df['Production']
```

```python
X_sampled, _, y_sampled, _ = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
# Spliting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_sampled, y_sampled, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Evaluating the Model
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
# Saving the model in a file
model_file_path = 'crop_production_model.joblib'
dump(model, model_file_path)
print("Model saved to:", model_file_path)
```

```python
# Plotinf actual value vs. predicted value
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Production")
plt.ylabel("Predicted Production")
plt.title("Actual vs. Predicted Production")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.residplot(x=y_pred, y=y_test - y_pred, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.xlabel('Predicted Production')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
